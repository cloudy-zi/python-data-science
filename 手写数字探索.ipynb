## 手写数字探索

>本文的数据集来自scikit-learn自带的手写数字集

```python
from sklearn.datasets import load_digits

digits = load_digits() ##加载数据集
digits.images.shape ##观察维度
```

```python
import matplotlib.pyplot as plt ##加载matpolib库

fig , axes = plt.subplots(10,10,figsize=(8,8),subplot_kw={'xticks':[],'yticks':[]},gridspec_kw = dict(hspace=0.1,wspace=0.1))
#fig 整个画布  axes每个小图
##10,10 创建10x10的网格图
#figsize 图的大小
#subplot_kw={'xticks':[],'yticks':[]} 给每个子图x,y设置参数 这里隐藏了
#gridspec_kw 子图之间的间距 只认字典

for i,ax in enumerate(axes.flat):  ##axes.flat 平铺成一维数组 enumerate 取i和ax图
    ax.imshow(digits.images[i],cmap='binary',interpolation='nearest')
    #在小图上画数字图片 cmap是配色 binary 是黑白配色   interpolation = ‘nearest'让像素清洗
    
    ax.text(0.05,0.05,str(digits.target[i]),transform=ax.transAxes,color = 'green')
    ##写出真实数字
```
